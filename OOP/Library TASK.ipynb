```python
class lib:
    def __init__(self,books={},user=[],password=[]):
        self.__user=user
        self.__password=password
        self.books=books
        
    def start(self):
        A = "start"
        while A != "stop":
            operation= input("Please enter operation (Register) or (Login) : ")
            if operation == "Register" or operation == "register" :
                lib().set_user()
            elif operation == "Login" or operation == "login" :
                lib().get_user()
            else :
                print ("invalid operation ")
                
            A = input ("do u wanna continue or stop ?")


    def set_user(self):
        x = input ("please enter user name : ")
        self.__user.append(x)
        y = input("please enter password : ")
        self.__password.append(y)
        print("Successfully Registered")
        self.books.update({x:0})
        
        
    def get_user(self):
        while True :
            x = input ("please enter user name : ")
            y = input("please enter password : ")
            if x in self.__user and y in self.__password :
                if self.__user.index(x) == self.__password.index(y):
                    print("Hello user :" , x)
                    lib().book(x)
                    break
                else :
                    print("invalid user name or password ")
            else :
                print("invalid user name or password ")
    
    def book(self,x):
        ask = input("u wanna loan a book or return a book ? ")
        if ask == "loan" :
            self.books[x]=self.books[x]+1
            print("ur current books is : " , self.books[x])
        elif ask == "return":
            self.books[x]=self.books[x]-1
            print("ur current books is : " , self.books[x])
            
lib().start()
```
